```r
# Finding Co-relation amongst variables

corr_simple <- function(x,sig=0.5){
#run a correlation and drop the insignificant ones
  corr <- cor(x)

  #prepare to drop duplicates and correlations of 1     
  corr[lower.tri(corr,diag=TRUE)] <- NA 

  #drop perfect correlations
  corr[corr == 1] <- NA 

  #turn into a 3-column table
  corr <- as.data.frame(as.table(corr))

  #remove the NA values from above 
  corr <- na.omit(corr) 

  #select significant values with Co-relation more than sig=0.5 
  corr <- subset(corr, abs(Freq) > sig) 

  #sort by highest correlation
  corr <- corr[order(-abs(corr$Freq)),] 

  #print table
  print(corr)

  #turn corr back into matrix in order to plot with corrplot
  mtx_corr <- reshape2::acast(corr, Var1~Var2, value.var="Freq")
  
  #plot correlations visually
  corrplot(mtx_corr, is.corr=FALSE, tl.col="black", na.label=" ")
}
corr_simple(X)
```

### 4.1 Linear Regression Model

```r
# Building Linear Regression model
linearMod <- lm(SalePrice ~ OverallQual, data=train)

#predict SalePrice
SalePrice_pred <- predict(linearMod, test)  # predict saleprice

summary(linearMod)
cat("Predicted values")
#Actual values Vs Predicted values
actuals_preds <- data.frame(cbind(actuals=test$SalePrice, predicteSalePrice=SalePrice_pred))  # make actuals_predicteds dataframe.
correlation_accuracy <- cor(actuals_preds) 
head(actuals_preds)

RMSE = mean((actuals_preds$actuals - actuals_preds$predicteSalePrice)^2) %>% sqrt()
cat("Root Mean Squared Error=",RMSE)
cat("\n")
cat("Correspondsing error rate = ",(0.07495/mean(train$SalePrice)*100))
```

### 4.2 Evaluation of Model

The overall quality of the linear regression fit can be assessed using the following three quantities, displayed in the model summary:

1. Residual Standard Error (RSE),
2. R-squared (R2) and adjusted R2,
3. F-statistic

In our example, using only OverallQual as the predictor variable, the Residual standard error is 0.07443. This means that the observed SalePrice values deviated from the predicted values by approximately 0.07443 units in average. This corresponds to an error rate of 0.07443/mean(train$SalePrice)=14.72%, which is good. 

The R2 measures how well the model fits the data. The higher the R2, the better the model. In our model, R2=0.6627

The adjustment in the “Adjusted R Square” value in the summary output is a correction for the number of x variables included in the predictive model. 
-An (adjusted) R2 that is close to 1 indicates that a large proportion of the variability in the outcome has been explained by the regression model.
-A number near 0 indicates that the regression model did not explain much of the variability in the outcome.

In our model, adj-R2=0.6624 which needs improvement.

The F-statistic gives the overall significance of the model. It assess whether at least one predictor variable has a non-zero coefficient. A large F-statistic will corresponds to a statistically significant p-value (p < 0.05). In our model, the F-statistic equal 2291 on 1 and 1166 DF, p-value: < 2.2e-16, which is highly significant.

### 4.3.2.2 Boosting Ensemble Model

```r
SalePrice_boost = gbm(SalePrice ~ ., data = train,  
                    n.trees = 5000, interaction.depth = 4, shrinkage = 0.01)
SalePrice_boost

#Predicting SalesPrice
SalePrice_pred = predict(SalePrice_boost, newdata = test, n.trees = 5000)

#RMSE error
RMSE = mean((SalePrice_pred - test$SalePrice)^2) %>% sqrt()
cat("Root Mean Squared Error=",RMSE)
cat("\n")
cat("Correspondsing error rate = ",(RMSE/mean(train$SalePrice)*100))

plot(SalePrice_pred, test$SalePrice,
     xlab = "Predicted", ylab = "Actual", 
     main = "Predicted vs Actual: Boosted Model, Test Data",
     col = "dodgerblue", pch = 20)
grid()
abline(0, 1, col = "darkorange", lwd = 2)
```

By using the Boosting Ensemble Model, and considering all variable effects on the target variable, our RSME = 0.03837 and error rate = 7.53%, which is less than the Multilinear Model with all the variables.




#### So, for this dataset, the Boosting Ensemble Model with all variables gave the highest accuracy of RSME = 0.03837 and error rate = 7.53%. Let's use this model to make our house price predictions.

#### Getting SalePrice From Model

```r
# Actual SalePrice dataframe
actual_SalePrice= data.frame(actual_SalePrice_norm=test$SalePrice)

# Predicted SalePrice dataframe
predicted_SalePrice= data.frame(predicted_SalePrice=SalePrice_pred)

# Finding Min and Max SalePrice values for denormalising
minval_SalePrice <- sapply(select(dataset,SalePrice),min)
maxval_SalePrice <- sapply(select(dataset,SalePrice),max)
minvec=minval_SalePrice 
maxvec=maxval_SalePrice 

#Denormalising function
denormalize <- function(x,minval,maxval) {
    x*(maxval-minval) + minval
}

# predicted SalePrice (log transformed values)
predicted_price=as.data.frame(Map(denormalize,predicted_SalePrice,minvec,maxvec))

# Actual SalePrice (log transformed values)
actual_price=as.data.frame(Map(denormalize,actual_SalePrice,minvec,maxvec))

#Actual SalePrice
actual_price= expm1(actual_price)

# predicted SalePrice
predicted_price= expm1(predicted_price)

# Finally, Comparison between Actual saleprice and Predicted saleprice
SalePrice_Comparison=data.frame(actual_SalePrice=actual_price$actual_SalePrice_norm,
                                predicted_SalePrice=predicted_price$predicted_SalePrice,
                               Difference_In_Percentage=((actual_price$actual_SalePrice_norm-predicted_price$predicted_SalePrice)/
                                           actual_price$actual_SalePrice_norm)*100)
SalePrice_Comparison
```
